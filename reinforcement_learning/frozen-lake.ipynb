```python
import gym
import numpy as np
from IPython.display import clear_output
import random
import time
```

```python
env = gym.make("FrozenLake-v1", is_slippery=False)
```

```python
# Set Hyperparam
learning_rate = 0.1
discount_rate = 0.9
num_episodes = 1000
max_steps_per_episode = 100
debug_model = True

exploration_rate = 1
max_exploration_rate = 1
min_exploration_rate = 0.01
exploration_decay_rate = 0.001
```

```python
max_iter = 1000
gamma = 0.9

num_states = env.observation_space.n
num_actions = env.action_space.n
state_prob = env.P

policy = np.zeros(num_states)
values = np.zeros(num_states)

for episode in range(max_iter):
    prev_value = np.copy(values)
    for state in range(num_states):
        q_value = []
        for action in range(num_actions):
            new_state_rewards = []
            for trans_prob, new_state, reward, done in state_prob[state][action]:
                new_state_rewards.append((trans_prob * (reward + gamma * prev_value[new_state])))
        
        q_value.append(sum(new_state_rewards))
        
    values[state] = max(q_value)
    
print(policy[0], episode)
```

```python
for state in range(num_states):
    q_sa = np.zeros(num_actions)
    for action in range(num_actions):
        for new_state in state_prob[state][action]:
            prob, new_state, reward, done = new_state
            q_sa[action] += (prob * (reward + gamma * values[new_state]))
    
    policy[state] = np.argmax(q_sa)
```
